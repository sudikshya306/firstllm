##Sentiment Analysis using transformers models

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!"
    ]
)
```

### Text generation model( like ChatGpt)

```python
# prompt: generator  =pipeline("text-generator")ge

generator  =pipeline("text-generation")
generator("In this course, we will teach you how to")
```

QA model


```python
qa = pipeline("question-answering")
qa(
    question="where do i work?",
    context="once upon a time in national college of engineering....",
)
```

### Language Trans;lator

```python
translator = pipeline("translation_en_to_de")
translator("I love programming language.")
```

### NER Models


```python
ner = pipeline("ner",model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am sudikshya and i am from kathmandu nepal")
```
